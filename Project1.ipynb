A matrix $A$ is normal if $A^{H}A = AA^{H} $


where:
$$
     A = U\Lambda U^H \text{ and } A^{H} = U\lambda^{H}U^{H}

the definition of the second norm is: 

$$
    \|A\|_2=\sqrt{\rho(A^{H}A)}= \sqrt{\rho(U\Lambda^H U^HU\Lambda U^H)} = \sqrt{\rho((U\Lambda^H\Lambda U^H)} = \sqrt{\rho(B)},   \text{ for  }  B = U\Lambda^{H}\Lambda U^H = U\Lambda_{B} U^H

the eigenvalues of B is given by the diagonal matrix $\Lambda_{B}$, where $\lambda_{B,i} = \lambda_{i}^2 $ for $i = 1,...,n$

$$
    \|B\|_2=\rho{(B)} = max_{1\leq i \leq n}\left | \lambda_{B,i}  \right | = max_{1\leq i \leq n}\left | \lambda_{i}^{2}  \right |

$$
   \|A\|_2 =  \left | \sqrt{\|B\|_2} \right | = max_{1\leq i \leq n}\left | \lambda_{i}  \right | = \rho{(A)} \text{      }\square

# Problem X



We assume that there exists a singular matrix $A + \delta A$ such that $(A + \delta A)x = 0$ for a $ x\neq 0$ ,  $x \in\mathbb{R}^{n}$ 

$$
    (I + \delta AA^{-1})x = 0A^{-1} = 0 \Leftrightarrow Ix = -\delta AA^{-1}x \Leftrightarrow I = -\delta AA^{-1}
$$

we can remove x since $x \neq 0 $, we now take the norm of both sides

$$
    1 = \|-\delta AA^{-1}| = \|\delta AA^{-1}| \leq \|A^{-1}\|\|\delta A\| 
$$

It is clear that for a singular $A + \delta A$ we get the inequality $\|\delta A\|\
\geq \|A^{-1}\|^{-1}$

Now we need to find out wether the fixed value $\|\delta A\|\ = \|A^{-1}\|^{-1} $ exists

A candidate for $ \delta A $ is $ \delta A = -\|A^{-1}\|^{-1}xy^{T}$ where

$$
    \|x\| = 1 \text{,  } \|y\|_* := max_{z\neq 0} \frac{|y^{T}z|}{\|z\|} = 1 \text{ and } \|A^{-1}\| = y^{T}A^{-1}x
$$


we use the definition of the matrix norm:

$$
        \|\delta A\| = max_{x \neq 0}\frac{\|\delta A z\|}{\|z\|} = max_{x \neq 0}\frac{\|-\|A^{-1}\|^{-1}xy^{T} z\|}{\|z\|} =max_{x \neq 0}\frac{\|-(y^{T}A^{-1}x)^{-1}xy^{T} z\|}{\|z\|}
$$

Since $-y^{T}A^{-1}x$ and $y^{T}z $ are scalars we get: 

$$
    \Rightarrow \|\delta A\| = max_{x \neq 0}\frac{|(y^{T}A^{-1}x)^{-1}y^{T} z|\|x\|}{\|z\|} = max_{x \neq 0}|(y^{T}A^{-1}x)^{-1} \text {, since } \|x|\ = 1 \text { and }  max_{z\neq 0} \frac{|y^{T}z|}{\|z\|} = 1
$$

Now we substitute back and get: 

$$
    \|\delta A\| = max_{x \neq 0}\|A^{-1}\|^{-1} = \|A^{-1}\|^{-1}
$$


We have proved that there exists a singular matrix $A + \delta A$ with the where the smallest possible norm $\|\delta A\| = \|A^{-1}\|^{-1}$

Finaly we want to prove that our matrix is singular, we look at

$$
    (A + \delta A)A^{-1}x = (I + \delta AA^{-1})x = (I -\|A^{-1}\|^{-1}xy^{T}\cdot \frac{\|A^{-1}\|}{xy^T})x = (I - I)x = 0x 
$$

we see that our guess for $A + \delta A$ is clearly singular and our proof holds $\square$
